### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

### Note
* Instructions have been included for each segment. You do not have to follow them exactly, but they are included to help you think through the steps.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()

```

## Player Count

* Display the total number of players


```python
total_player=purchase_data["SN"].unique()
total_players=len(total_player)
print ("Total Players")
print(total_players)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items=len(purchase_data["Item ID"].unique())
average_price=round(purchase_data.Price.mean(),2)
total_pruchases=purchase_data["Purchase ID"].count()
total_revenue=purchase_data.Price.sum()
summary_dict=([{"Number of Unique Items":items,
                      "Average Price":average_price,
                      "Number of Purchases":total_pruchases,
                      "Total Revenue":total_revenue}])
summary = pd.DataFrame(summary_dict)
summary["Average Price"]=summary["Average Price"].map("${:.2f}".format)
summary["Total Revenue"]=summary["Total Revenue"].map("${:,.2f}".format)
summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
male_players=purchase_data[purchase_data["Gender"]=="Male"]
count_male=len(male_players["SN"].unique())
male_percent=round((count_male/total_players)*100,2)

female_players=purchase_data[purchase_data["Gender"]=="Female"]
count_female=len(female_players["SN"].unique())
female_percent=round((count_female/total_players)*100,2)

other=total_players-count_male-count_female
other_percent=round((other/total_players)*100,2)

Gender_demo=pd.DataFrame({"Total Count":[count_male,count_female,other],
                      "Percentage of Players":[male_percent,female_percent,other_percent]},index = ['Male','Female','Other / Non-Disclosed'] ,
                    columns=['Total Count','Percentage of Players'])
Gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_group=purchase_data.groupby("Gender")
average=round(gender_group.Price.mean(),2)
count=gender_group.Gender.count()
total_purchase=gender_group.Price.sum()
unique_person=gender_group["SN"].nunique()
purchase_by_person=round(total_purchase/unique_person,2)
purchase_analysis=pd.DataFrame({"Purchase Count":count,"Average Purchase Price":average,"Total Purchase Value":total_purchase,"Avg Total Purchase per Person":purchase_by_person})
purchase_analysis=purchase_analysis.reset_index()
purchase_analysis["Average Purchase Price"]=purchase_analysis["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis["Total Purchase Value"]=purchase_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis["Avg Total Purchase per Person"]=purchase_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
purchase_analysis
```

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins,labels=group_names)
age_demog = purchase_data.groupby("Age Group")
age_demog_count =age_demog["SN"].nunique()
age_demog_perc = round((age_demog_count/total_players)*100,2)
age_demog_summary = pd.DataFrame({"Total Count":age_demog_count,"Percentage of Players":age_demog_perc})
age_demog_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_age_count=age_demog.Age.count()
purchase_age_average=round(age_demog.Price.mean(),2)
purchase_age_total=age_demog.Price.sum()
unique_age_person=age_demog["SN"].nunique()
purchase_by_person_age=round(purchase_age_total/unique_age_person,2)
purchase_data_age=pd.DataFrame({"Purchase Count":purchase_age_count,"Average Purchase Price":purchase_age_average,"Total Purchase Value":purchase_age_total,"Avg Total Purchase per Person":purchase_by_person_age})
purchase_data_age=purchase_data_age.reset_index()
purchase_data_age["Average Purchase Price"]=purchase_data_age["Average Purchase Price"].map("${:.2f}".format)
purchase_data_age["Total Purchase Value"]=purchase_data_age["Total Purchase Value"].map("${:,.2f}".format)
purchase_data_age["Avg Total Purchase per Person"]=purchase_data_age["Avg Total Purchase per Person"].map("${:.2f}".format)
purchase_data_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grouped_SN = purchase_data.groupby("SN")

pur_count_SN = grouped_SN["Purchase ID"].count()
total_pur_SN = grouped_SN["Price"].sum()
avg_pur_price_SN = total_pur_SN/pur_count_SN

summary_SN = pd.DataFrame({"Purchase Count":pur_count_SN,
                               "Average Purchase Price":avg_pur_price_SN,
                               "Total Purchase Value":total_pur_SN})
summary_top_SN = summary_SN.sort_values("Total Purchase Value",ascending=False)
summary_top_SN["Average Purchase Price"] = summary_top_SN["Average Purchase Price"].map("${:.2f}".format)
summary_top_SN["Total Purchase Value"] = summary_top_SN["Total Purchase Value"].map("${:.2f}".format)
summary_top_SN.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grouped_item = purchase_data.groupby(["Item ID","Item Name","Price"])
pur_count_item = grouped_item["Purchase ID"].count()
total_pur_item = grouped_item["Price"].sum()
summary_item = pd.DataFrame({"Purchase Count":pur_count_item,"Total Purchase Value":total_pur_item})
summary_item=summary_item.reset_index()
summary_top_item = summary_item.sort_values("Purchase Count",ascending=False)
summary_top_item["Price"] = summary_top_item["Price"].map("${:.2f}".format)
summary_top_item["Total Purchase Value"] = summary_top_item["Total Purchase Value"].map("${:.2f}".format)
summary_top_item.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
summary_top_item = summary_item.sort_values("Total Purchase Value",ascending=False)
summary_top_item["Price"] = summary_top_item["Price"].map("${:.2f}".format)
summary_top_item["Total Purchase Value"] = summary_top_item["Total Purchase Value"].map("${:.2f}".format)
summary_top_item.head()
```
